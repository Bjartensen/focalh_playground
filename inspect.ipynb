```python
import ROOT
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import pandas as pd
import focal
```

# Simple script to inspect converted files

```python
filename = "/home/bjartur/workspace/focalh_data_transformer/data/converted/E300_P5_N1000.root"
```

```python
def test_file(filename):
    tfile = ROOT.TFile(filename, "READ")
    ttree = tfile.Get("EventsTree")
    foc = focal.FocalH()

    rows=4
    cols=4
    fig, ax = plt.subplots(nrows=rows, ncols=cols, figsize=(10,10))
    ax = ax.flatten()

    for i in range(rows*cols):
        foc.heatmap(ttree, i, ax[i])
    tfile.Close()

test_file(filename)
```
